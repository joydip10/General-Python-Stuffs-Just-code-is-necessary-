```python
day=[1,2,3,4,5,6,7]
max_temp=[23,24,25,24,23,23,24]
min_temp=[21,23,23,22,23,20,23]
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.plot(day,max_temp,label='max_temp')
plt.plot(day,min_temp,label='min_temp')
plt.legend(loc='best',shadow=True,fontsize='small')
plt.grid()
```

```python
#bar
import numpy as np
stocks=['DSE','Shangai','USTrade','Hongkong','Thailand','Russia','Swedish']
close=[23,24.5,21.2,20,24,26.7,25.2]
profits=['good','bad','bad','good','good','good','bad']
profits_num=[]
for i in profits:
    if i =='good':
        profits_num.append(2)
    else:
        profits_num.append(1)
profits_num=np.array(profits_num)
xpos=np.arange(len(stocks))
plt.xticks(xpos,stocks)
plt.xlabel('stocks')
plt.ylabel('Value')
plt.title('Money Exchange')
plt.bar(xpos-0.2,close,label='stockreturns',width=0.7)
plt.bar(xpos+0.2,profits_num,label='profits/loss',width=0.5)
plt.legend(loc='best',shadow=True,fontsize='small')
```

```python
#bar_horizontal_fashion
import numpy as np
stocks=['DSE','Shangai','USTrade','Hongkong','Thailand','Russia','Swedish']
close=[23,24.5,21.2,20,24,26.7,25.2]
profits=['good','bad','bad','good','good','good','bad']
profits_num=[]
for i in profits:
    if i =='good':
        profits_num.append(2)
    else:
        profits_num.append(1)
profits_num=np.array(profits_num)
ypos=np.arange(len(stocks))
plt.yticks(ypos,stocks)
plt.ylabel('stocks')
plt.xlabel('Value')
plt.title('Money Exchange')
plt.barh(ypos-0.2,close,label='stockreturns',color='yellow')
plt.barh(ypos+0.2,profits_num,label='profits/loss',color='red')
plt.legend(loc='best',shadow=True,fontsize='small')
```

```python
#histogram with single sample
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,rwidth=0.8,bins=[80,100,125,150],color='green',histtype='step')#we can also use the bin parameter
plt.xlabel('Blood Sugar Label')
plt.ylabel('Number of patients')
plt.title('Diabetic histogram')
```

```python
#histogram with multiple sample
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel('number of patients')
plt.ylabel('sugar level')
plt.title('Diabetic data')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],color=['red','green'],rwidth=0.95,label=['Men','Women'],orientation='horizontal')
plt.legend()
```
