```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5]
y=[23,24,25,21,23]
plt.plot(x,y,color='green',linewidth=2,linestyle='dashed',marker='D',markersize=16,alpha=0.5)
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('plotting Day by Temperature')
```

```python
days=[1,2,3,4,5,6,7]
max_t=[31,32,33,32,33,32,34]
min_t=[30,29,30,31,32,31,33]
avg_t=[30.5,30,31,31,31,32,33]
plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Average')
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
plt.grid()
```

```python
company=['Google','Faceboook','Amazon','Twitter','Apple']
revenue=[8.5,6.7,8.1,5.6,8.3]
profit=[20,19,22,15,20]
#plt.bar(company,revenue)
xpos=np.arange(len(company))
plt.xticks(xpos,company)
plt.bar(xpos+0.2,revenue,label='revenue',width=0.3)

plt.xlabel('Company')
plt.ylabel('Revenues')
plt.title('Chart')

plt.bar(xpos-0.3,profit,label='profit',width=0.4)
plt.legend(bbox_to_anchor=(1.3,1),loc=1)
```

```python
company=['Google','Faceboook','Amazon','Twitter','Apple']
revenue=[8.5,6.7,8.1,5.6,8.3]
profit=[20,19,22,15,20]
#plt.bar(company,revenue)
ypos=np.arange(len(revenue))
plt.yticks(ypos,company)
plt.barh(ypos+0.2,revenue,label='revenue')

plt.ylabel('Company')
plt.xlabel('Revenues')
plt.title('Chart')

plt.barh(ypos-0.3,profit,label='pofit')
plt.legend(loc='best')
```
