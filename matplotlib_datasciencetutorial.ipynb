```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
print(x)
print(y)
```

```python
plt.plot(x,y,linestyle='dashed',marker='+',color='red')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color='red')
plt.subplot(1,2,2)
plt.plot(y,x,color='blue')
plt.show()
```

```python
#object oriented
```

```python
fig=plt.figure()


axes=fig.add_axes([0.1,0.2,0.9,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Graph')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(x,y,linestyle='dashed')
axes2.plot(y,x,linestyle='dashed')

axes2.set_title('smaller plot')
axes1.set_title('Larger plot')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)
plt.tight_layout()

#for current_ax in axes:
#    current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('Graph')
axes[0].set_xlabel('xlabel')
axes[0].set_ylabel('ylabel')


axes[1].plot(y,x)
axes[1].set_title('Reversed Graph')
axes[1].set_xlabel('xlabel')
axes[1].set_ylabel('ylabel')
```

```python
#figure size and DPI

fig=plt.figure(figsize=(8,6),dpi=100)
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,7),dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='x vs x**2')
axes.plot(x,x**3,label='x vs x**3')
axes.legend(loc='best',shadow=True,fontsize='large')
axes.legend(bbox_to_anchor=(1,1))
```

```python
fig=plt.figure(figsize=(8,5),dpi=200)

ax=fig.add_axes([0,0,1,1])
ax1=fig.add_axes([0.1,0.5,0.3,0.3])
ax1.plot(x,y,color='g',linestyle='dashed',marker='+')
ax1.set_xlim([0,1])
ax1.set_ylim([0,2])
ax.plot(x,y,color='g',linestyle='dashed',marker='+')
```
