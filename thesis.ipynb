```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from pylab import rcParams
%matplotlib inline

density_for_o_o1=[]
assortivity_o_o1=[]
clustering_o_o1=[]

density_for_o_o5=[]
assortivity_o_o5=[]
clustering_o_o5=[]

density_for_o_1=[]
assortivity_o_1=[]
clustering_o_1=[]

year=[2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
dataframe_list=[data2000,data2001,data2002,data2003,data2004,data2005,data2006,data2007,data2008,data2009,data2010,data2011,data2012,data2013,data2014,data2015,data2016,data2017,data2018]

def findproperties(pvalue):
    l=[[]for i in range(21)]
    l=pvalue    
    grapharray=np.array(l)       
    g=nx.DiGraph(grapharray)
    edges=(len(list(g.edges))-21)//2
    network_density=((21*(21-1))/2)/(edges/2)
    network_assortivity=nx.degree_assortativity_coefficient(g)
    network_clustering=nx.average_clustering(g)
    return network_density,network_assortivity,network_clustering

#0.01
for i in pvalue_o_o1:
    d,a,c=findproperties(i)
    density_for_o_o1.append(d)
    assortivity_o_o1.append(a)
    clustering_o_o1.append(c)

#0.05
for i in p_value_o_o5:
    d,a,c=findproperties(i,0.05)
    density_for_o_o5.append(d)
    assortivity_o_o5.append(a)
    clustering_o_o5.append(c)

#0.1
for i in p_value_o_1:
    d,a,c=findproperties(i)
    density_for_o_1.append(d)
    assortivity_o_1.append(a)
    clustering_o_1.append(c)

width_in_inches = 10
height_in_inches = 4
dots_per_inch = 100

plt.figure(
    figsize=(width_in_inches, height_in_inches),
    dpi=dots_per_inch)    
    
plt.axis([2000,2018,0,10])
plt.xticks([i for i in range(2000,2019)])
plt.xlabel('year')
plt.ylabel('Network Density')
plt.title('Changes in network density due to change in p-value')
plt.plot(year,density_for_o_o1,color='red',label='pvalue=0.01')
plt.plot(year,density_for_o_o5,color='green',label='pvalue=0.05')
plt.plot(year,density_for_o_1,color='blue',label='pvalue=0.1')
plt.legend(loc='best',shadow=True)
rcParams['figure.figsize'] = 5, 10
plt.tight_layout()
```
