```python
#matrix plot
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
import pandas as pd
l=list(tips.total_bill)
m=list(tips.tip)
n=list(tips['size'])
df=pd.DataFrame(data=[l,m,n])
df=df.T
df.rename(columns={0:'total_bill',1:'tips',2:'size'},inplace=True)
df.set_index('size',inplace=True)
corr=df.corr()
corr
```

```python
#heatmap
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
flights=flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(flights,annot=False,cmap='magma',linewidth=1,linecolor='white',vmin=20,vmax=400)
flights
```

```python
sns.clustermap(flights,cmap='coolwarm',standard_scale=5)
```
